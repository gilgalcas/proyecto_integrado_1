```python
# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos desde GitHub
# Reemplazar con tu URL de GitHub raw
url = 'https://raw.githubusercontent.com/gilgalcas/proyecto_integrado_1/main/ventas_tecnologia.csv.xlsx'
df = pd.read_excel(url)

# Check the actual column names in your DataFrame
print(df.columns)

# Replace 'Mes' with the actual name of your month column
# For example, if the column is named 'Unnamed: 0', use:
month_column = df.columns[0] # Assuming 'Mes' is the first column

# **Convert the month column to string type**
# **Ensure all values in the month column are strings BEFORE plotting**
# **This addresses the issue if your month column is a mix of numbers and strings:**

# Applying to_string() to each element individually
# **Explicitly convert to string using str() before applying astype()**
df[month_column] = df[month_column].apply(str).astype(str)  # Convert each element to string

# **Force matplotlib to treat the x-axis as categorical:**
# **This is the key change to prevent the TypeError.**
fig, ax = plt.subplots(figsize=(15, 8))  # Create figure and axes

for columna in df.columns[1:]:  # Excluimos la columna 'Mes'
    # Ensure both x and y values are numeric
    # This handles the case where the 'columna' data contains non-numeric values
    x_values = df[month_column]
    y_values = pd.to_numeric(df[columna], errors='coerce')  # Convert to numeric, handling errors

    # Filter out NaN values to avoid plotting errors
    mask = ~y_values.isna()  # Create a mask for non-NaN values
    ax.plot(x_values[mask], y_values[mask], marker='o', label=columna)  # Plot only valid data


# Set x-axis type to categorical using ax.set_xticks and ax.set_xticklabels
ax.set_xticks(df[month_column].unique())  # Set unique month values as ticks
ax.set_xticklabels(df[month_column].unique(), rotation=45)  # Set tick labels and rotation

ax.set_title('Ventas de Productos Tecnológicos 2023')
ax.set_xlabel('Mes')
ax.set_ylabel('Unidades Vendidas')
ax.legend()
ax.grid(True)
fig.tight_layout()  # Use fig.tight_layout() instead of plt.tight_layout()
plt.show()
```
